# 合并n个升序链表

给你一个链表数组，每个链表都已经按升序排列。

请你将所有链表合并到一个升序链表中，返回合并后的链表。

示例 1：

```
输入：lists = [[1,4,5],[1,3,4],[2,6]]
输出：[1,1,2,3,4,4,5,6]
解释：链表数组如下：
[
  1->4->5,
  1->3->4,
  2->6
]
将它们合并到一个有序链表中得到。
1->1->2->3->4->4->5->6
```

示例 2：

```
输入：lists = []
输出：[]
```

示例 3：

```
输入：lists = [[]]
输出：[]
```

````python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


class MyNode:
    def __init__(self, nd):
        self.val = nd.val 
        self.nd = nd

    def __lt__(self, node):
        return self.val < node.val

    def __iter__(self):
        yield self.val
        yield self.nd

import heapq as hq

class Solution:
    def mergeKLists(self, lists: List[ListNode]) -> ListNode:
        heap = []
        for l in lists:
            if l is None:
                continue
            hq.heappush(heap, MyNode(l))

        root = ListNode(-1, None)
        p = root

        while len(heap) > 0:
            val, l = hq.heappop(heap)
            p.next = ListNode(val)
            if l.next != None:
                hq.heappush(heap, MyNode(l.next))
            p = p.next

        return root.next
````
